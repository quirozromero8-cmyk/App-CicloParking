```python
import pandas as pd

df = pd.read_csv("Cicloparqueadero.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df['CLASECICLOP'] = df['CLASECICLOP'].replace({'F': 1,'I': 2})
df['CLASECICLOP'] = df['CLASECICLOP'].astype(int)
class_df = df['CLASECICLOP'].value_counts()
print(class_df)
```

```python
df['SELLOCICLOP'] = df['SELLOCICLOP'].replace(2, 3)
sello_df = df['SELLOCICLOP'].value_counts()
print(sello_df)
```

```python
Quality_df = df.copy()
Quality_df['Categoria'] = Quality_df['CLASECICLOP'] + Quality_df['SELLOCICLOP']

Quality_df["Categoria"] = (
    Quality_df["CLASECICLOP"] + Quality_df["SELLOCICLOP"]
).apply(lambda x: "Oro" if x == 4 else "Plata")

print(Quality_df['Categoria'])
```

```python
#Suma de las distancias Columna X y Y
distancia_df = df.copy()
distancia_df["Distancia"] = df["X"].abs() + df["Y"].abs()
print(distancia_df['Distancia'])
```

```python
df_parqueaderos = pd.DataFrame({
    "X": df["X"],
    "Y":  df["Y"],
    'Distancia': distancia_df['Distancia'],
    'Nombre': df['NOMBRECICLOP'],
    'Direccion': df['DIRECCIONCICLOP'],
    'Horarios': df['HORARIOCICLOP'],
    'Categoria': Quality_df['Categoria'],
    })

print(df_parqueaderos)
```
